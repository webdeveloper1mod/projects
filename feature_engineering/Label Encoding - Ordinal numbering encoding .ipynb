```python
import pandas as pd
import datetime
```

```python
#for example: students grade A,B,C or Fail
# I create a list of dates with 20 days difference from today 
# and then transform it into a dataframe
df_base = datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(0,200)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
#extract the week day name 
df['day_of_week'] = df['day'].dt.day_name()
df.head()
```

```python
# Engineering categorical variable by ordinal number replacement
```

```python
weekday_map = {"Monday": 1,
               "Tuesday": 2,
               "Wednesday": 3,
               "Thursday": 4,
               "Friday": 5,
               "Saturday": 6,
               "Sunday": 7
}
df['day_ordinal'] = df.day_of_week.map(weekday_map)
df.head(20)
```
